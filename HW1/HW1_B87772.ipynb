ex2: I ended up by choosing the merge sort algorithm. Here below you can find the code of it.

```python
def mergeSort(mlist):
    if len(mlist)>1:
        mid = len(mlist)//2
        lefthalf = mlist[:mid]
        righthalf = mlist[mid:]

        mergeSort(lefthalf)
        mergeSort(righthalf)
        
        return merge(lefthalf, righthalf, mlist)
    
def merge(lefthalf, righthalf, mlist):
        i=0
        j=0
        k=0
        while i < len(lefthalf) and j < len(righthalf):
            if lefthalf[i] < righthalf[j]:
                mlist[k]=lefthalf[i]
                i=i+1
            else:
                mlist[k]=righthalf[j]
                j=j+1
            k=k+1

        while i < len(lefthalf):
            mlist[k]=lefthalf[i]
            i=i+1
            k=k+1

        while j < len(righthalf):
            mlist[k]=righthalf[j]
            j=j+1
            k=k+1
```

ex3: Phyton's the most used sorting function is Timsort.
Timsort is a hybrid sorting algorithm, derived from merge sort and insertion sort, designed to perform well on many kinds of real-world data. It was invented by Tim Peters in 2002 for use in the Python programming language. The algorithm finds subsets of the data that are already ordered, and uses the subsets to sort the data more efficiently. This is done by merging an identified subset, called a run, with existing runs until certain criteria are fulfilled. Timsort has been Python's standard sorting algorithm since version 2.3. The estimated Big O of this algorithm is n∗log(n)/45.

```python
def increasing(n):
    measures = []
    f_x = []
    for j in range(1, n):
        randomArr = numpy.array(numpy.random.randint(0,9223372036854775807,j, dtype=numpy.int64)).tolist()
        startTime = time.time() 
        randomArr.sort()
        duration = time.time() - startTime
        measures.append(duration)
        f_x.append((numpy.log(j) * duration * j)/45)
    plott.title("Analysis of .sort complexity")
    plott.ylabel('Seconds')
    plott.xlabel('Number of integers')

    plott.plot(measures, label='.sort()')
    plott.plot(f_x, label='(log(n)*n)')
    plott.legend()
    plott.show()
increasing(1000)
```

ex5: for \begin{equation*} f(n) = 4^3 + 0.5n , g(n) = n^3 \end{equation*} we should prove that f(n) = Theta(g(n)) by finding suitable c1, c2 and n0. Firstly we should prove that f(n)∈O(g(n))∩Ω(g(n))

O(g(n)):

f(n)∈O(g(n))⟺0 ≤ f(n) ≤ c1∗g(n)  for any  n≥n0 with  c1>0,n0>0. It means that \begin{equation*} 0≤4n^3+0.5n≤c1∗n^3 
\end{equation*}


This assertion is only true when  c1≥4.5

The limit of the comparison should then be positive to prove that in this case, 
g(n) is always bigger than f(n) for any $n > n_0$ :

$\lim_{n \to \infty} \frac{4.5 n^{3}}{4 n^{3} + 0.5 n} = \lim_{n \to \infty} \frac{4.5}{4 + \frac{0.5}{n^{2}}} = 1.125 \Longrightarrow positive$


For any $n > n_0$ , this term is positive. As such, $f(n) \in O(g(n)).$

$\Omega(g(n))$:

$f(n) \in \Omega(g(n)) \Longleftrightarrow 0 \leq c_2 * g(n) \leq f(n)$ for any $n \geq n_0$ with $c_2 > 0, n_0 > 0$ that means that $0 \leq c_2 * n^{3} \leq 4n³ + 0.5n.$ This assertion is only true when $c_2 \leq 4.$

The limit of the comparison should then be zero to prove that in this case, g(n) is always smaller than f(n) for any $n > n_0$ :

$\lim_{n \to \infty} \frac{4 n^{3} + 0.5 n}{4 n^{3}} = \lim_{n \to \infty} \frac{0.5}{4n^{2}} = \lim_{n \to \infty} \frac{0.125}{n^{2}} = 0$


For any  $n > n_0$, this term is zero. As such, $f(n) \in \Omega(g(n)).$

$\Theta(g(n))$

We know now that $f(n) \in O(g(n)) \cap \Omega(g(n))$, thus, $f(n) \in \Theta(g(n)).$
